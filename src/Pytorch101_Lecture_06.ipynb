# Backward Propagation in Pytorch

Let us consider the following Computational Graph

ols.png

How do we minimize such a loss?

We need to calculate the local gradients at each node inside the computational graph, namely:
$$\frac{\partial \text{Loss}}{\partial l}, \quad \frac{\partial l}{\partial \hat{y}}, \quad \frac{\partial \hat{y}}{\partial w} $$
and then those are used to get
$$\frac{\partial \text{Loss}}{\partial w} $$
via backpropagation.

```python
import torch
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)
```

```python
w = torch.tensor(1.0, requires_grad=True)
```

```python
y_hat = w * x
loss = (y_hat - y) ** 2
```

```python
print(f"Forward Step: {loss}")
```

```python
loss.backward()
print(f'Backward Step: {w.grad}')
```

```python
with torch.no_grad():
  w -= 0.01 * w.grad
  print(w) # 1.02
print(w.grad) # -2
```
